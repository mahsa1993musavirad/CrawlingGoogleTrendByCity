```python
import pandas as pd      
from datetime import date
from pytrends.request import TrendReq
maindf  = pd.DataFrame()
tempdf = pd.DataFrame()
pytrend = TrendReq()
pytrend.build_payload(kw_list=[allcities[0]],timeframe='2019-01-1 2019-03-31', geo='')

maindf = pytrend.interest_over_time()
maindf = maindf.drop(columns='isPartial')

for index in range(1, len(allcities)):


    pytrend.build_payload(kw_list=[allcities[index]] ,timeframe='2019-01-1 2019-03-31', geo='')

    tempdf = pytrend.interest_over_time()
    #print(tempdf.columns)
    try:
        
        tempdf = tempdf.drop(columns='isPartial')
    except:
        pass
    #print(tempdf)
    maindf = pd.concat([maindf, tempdf], axis=1)
#print(maindf)
```
